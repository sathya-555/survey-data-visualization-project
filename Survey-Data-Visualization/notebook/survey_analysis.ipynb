```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("survey_data.csv")
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Learning_Mode", data=df)
plt.title("Preferred Learning Mode")
plt.xlabel("Learning Mode")
plt.ylabel("Number of Respondents")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df["Gender"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90)
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[["Satisfaction", "Internship_Rating"]].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Between Ratings")
plt.show()
```

```python
## Key Insights
##- Online learning is the most preferred mode.
##- Majority of interns rated satisfaction above 4.
##- Satisfaction and internship rating show positive correlation.
```
